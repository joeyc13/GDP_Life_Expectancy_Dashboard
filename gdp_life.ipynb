```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
if 'data' not in pn.state.cache.keys():
    
    df = pd.read_csv("all_data.csv")
    
    pn.state.cache['data'] = df.copy()

else:
    
    df = pn.state.cache['data']
```

```python
idf = df.interactive()
```

```python
# idf
```

```python
countries = ['Chile', 'China', 'Germany', 'Mexico', 'United States of America', 'Zimbabwe']

country_radio_group = pn.widgets.RadioBoxGroup(name='CountryRadioBoxGroup', options=countries, inline=False)

country_radio_group
```

```python
country_radio_group.value
```

```python
line_pipeline = idf[idf.Country == country_radio_group]

# gdp_year_line_pipeline
```

```python
gdp_plot = line_pipeline.hvplot.line(x='Year',
                      y='GDP',
                      size=80, height=500, width=500)

gdp_plot
```

```python
life_expect_plot = line_pipeline.hvplot.line(x='Year',
                                             y='Life expectancy at birth (years)',
                                             size=80, height=500, width=500)

life_expect_plot
```

```python
template = pn.template.FastListTemplate(
    title='GDP v Life Expectancy Dashboard',
    sidebar=[pn.pane.Markdown("# Comparing the Relationship Between GDP and Life Expectancy"),
             pn.pane.Markdown("### The purpose of this dashboard is to help users visualize the relationship between GDP and life expectancy for various countries around the world."),
             pn.pane.PNG('earth.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Setings"),
             country_radio_group],
    main=[pn.Row(gdp_plot.panel(width=1000, margin=(0,25)),
                 life_expect_plot.panel(width=1000, margin=(0,25)), sizing_mode='stretch_width')],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)

template.show()
# template.servable();
```
